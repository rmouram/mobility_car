### 0.1 Helper Functions

```python
def mean_absolute_percentage_error(y, yhat):
    return np.mean(np.abs((y-yhat)/y))

def ml_error(model_name, y, yhat):
    mae = mean_absolute_error(y, yhat)
    mape = mean_absolute_percentage_error(y, yhat)
    rmse = np.sqrt(mean_squared_error(y, yhat))
    
    return pd.DataFrame({'Model Name': model_name,
                         'MAE': mae,
                         'MAPE': mape,
                         'RMSE': rmse}, index=[0])
```

## **2.0 Feature Engineering**

```python
df2 = df1.copy()
```

```python
# Desmembrando a coluna versao: 

# Motor 
df2['motor'] = df2['versao'].str.extract('(\d\.\d)')
df2['motor'].fillna(0, inplace=True)

# Turbo
df2['turbo'] = df2['versao'].str.extract('(TURBO)')
df2['turbo'] = df2['turbo'].apply(lambda x: x.replace('TURBO', '1') if pd.notnull(x) else '0')
df2['turbo'] = df2['turbo'].astype(int)

# Combustível
df2['combustivel'] = df2['versao'].str.extract('(FLEX|DIESEL|ALCOOL|GASOLINA|HÍBRIDO|ELÉTRICO|HYBRID)')
df2['combustivel'].fillna('FLEX', inplace=True)
df2['motor'] = df2['motor'].astype(float)

# Tração
df2['4x4'] = df2['versao'].str.extract('(4X4|4WD|AWD)')

df2['4x4'] = df2['4x4'].apply(lambda x: x.replace('4X4', '1') if pd.notnull(x) else x)
df2['4x4'] = df2['4x4'].apply(lambda x: x.replace('4WD', '1') if pd.notnull(x) else x)
df2['4x4'] = df2['4x4'].apply(lambda x: x.replace('AWD', '1') if pd.notnull(x) else x)
df2['4x4'].fillna('0', inplace=True)
df2['4x4'] = df2['4x4'].astype(int)
```

#### **5.3 Encoding**

```python
# Label Encoding:

le = LabelEncoder()
df5['cambio'] = le.fit_transform(df5['cambio'])
df5['tipo'] = le.fit_transform(df5['tipo'])
df5['blindado'] = le.fit_transform(df5['blindado'])
df5['cor'] = le.fit_transform(df5['cor'])
df5['tipo_vendedor'] = le.fit_transform(df5['tipo_vendedor'])
df5['tipo_anuncio'] = le.fit_transform(df5['tipo_anuncio'])
df5['combustivel'] = le.fit_transform(df5['combustivel'])

# Target Encoding (ideal para muitas variáveis categóricas): marca, modelo, versao, cidade_vendedor, estado_vendedor

# marca (frequency encode)
fe_marca = df5.groupby('marca').size() / len(df5)
df5.loc[: , 'marca'] = df5['marca'].map(fe_marca)

# modelo (frequency encode)
fe_modelo = df5.groupby('modelo').size() / len(df5)
df5.loc[: , 'modelo'] = df5['modelo'].map(fe_modelo)
#pickle.dump(fe_modelo, open('../parameter/fe_modelo_scaler.pkl', 'wb'))

# versao (frequency encode)
fe_versao = df5.groupby('versao').size() / len(df5)
df5.loc[: , 'versao'] = df5['versao'].map(fe_versao)

# cidade_vendedor (target encode)
target_encode_cidade_vendedor = df5.groupby('cidade_vendedor')['preco'].mean()
df5.loc[:,'cidade_vendedor'] = df5['cidade_vendedor'].map(target_encode_cidade_vendedor)

# estado_vendedor (target encode)
target_encode_estado_vendedor = df5.groupby('estado_vendedor')['preco'].mean()
df5.loc[:,'estado_vendedor'] = df5['estado_vendedor'].map(target_encode_estado_vendedor)

# modelo (target encode)
# target_encode_modelo = df5.groupby('modelo')['preco'].mean()
# df5.loc[:,'modelo'] = df5['modelo'].map(target_encode_modelo)

# versao (target encode)
# target_encode_versao = df5.groupby('versao')['preco'].mean()
# df5.loc[:,'versao'] = df5['versao'].map(target_encode_versao)

# marca (target encode)
# target_encode_marca = df5.groupby('marca')['preco'].mean()
# df5.loc[:,'marca'] = df5['marca'].map(target_encode_marca)
#pickle.dump(target_encode_marca, open('../parameter/target_encode_marca_scaler.pkl', 'wb'))
```

### **6.1 Features Importance**

```python
X = df5.drop('preco', axis=1)
y = df5['preco'].copy()

x_train, x_val, y_train, y_val = ms.train_test_split(X, y, test_size=0.20)
```

```python
x_train.shape
```

```python
# model definition
forest = ExtraTreesRegressor(n_estimators = 250, random_state=0, n_jobs=-1)

#data preparation
x_train_n = df6.drop(['preco', 'num_fotos'], axis=1)
y_train_n = y.values
forest.fit(x_train_n, y_train_n)
```

```python
importances = forest.feature_importances_
std = np.std([tree.feature_importances_ for tree in forest.estimators_], axis=0)
indices = np.argsort(importances)[::-1]

# print the feature ranking
print("Feature Ranking:")
df = pd.DataFrame()
for i, j in zip(x_train_n, forest.feature_importances_):
    aux=pd.DataFrame({'feature':i, 'importance':j}, index=[0])
    df = pd.concat([df, aux], axis=0)

print(df.sort_values('importance', ascending=False))

# Plot the impurity-based feature importances of the forest
plt.figure()
plt.title("Feature_importances")
plt.bar(range(x_train_n.shape[1]), importances[indices], color='r', yerr=std[indices], align='center')
plt.xticks(range(x_train_n.shape[1]), indices)
plt.xlim((-1, x_train_n.shape[1]))
plt.show()
```
